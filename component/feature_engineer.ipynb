```python
import pandas as pd
```

```python
label = pd.read_csv("data/label.csv")
commobd = pd.read_csv("data/raw_commorbidy.csv")
labevents = pd.read_csv("data/raw_labevents.csv")
procdevnts = pd.read_csv("data/raw_procedureevents.csv")
patientsinfo = pd.read_csv("data/raw_patientsinfo.csv")
```

```python
# Unstack location & location category
location = procdevnts[['hadm_id', 'location', 'locationcategory']]
location['location'] = location['location'].astype('str')
location['location'] = location['location'].map(lambda x: x.replace('Right ', '').replace('Left ', '').replace('R ', '').replace('L ', '').replace('.', '').rstrip())
location['locationcategory'] = location.locationcategory.astype('str')

# Merge unstacked df
temp1 = location.groupby(['hadm_id', 'location']).size().unstack(fill_value=0, level=-1).reset_index()
temp2 = location.groupby(['hadm_id', 'locationcategory']).size().unstack(fill_value=0, level=-1).reset_index()
merge_temp = pd.merge(temp1, temp2, on='hadm_id', how='outer').fillna('missing')
merge_temp
```

```python
# Same operation with manally classified location
location2 = location.copy()
location2['location'] = location2['location'].map(lambda x: x.replace('Right ', '').replace('Left ', '').replace('R ', '').replace('L ', '').replace('.', '').rstrip().replace('Ventricular', 'ICP monitors').replace('Lumbar', 'ICP monitors').replace('Intraparenchymal', 'ICP monitors').replace('Ulnar', 'Radial'))

temp12 = location2.groupby(['hadm_id', 'location']).size().unstack(fill_value=0, level=-1).reset_index()
temp22 = location2.groupby(['hadm_id', 'locationcategory']).size().unstack(fill_value=0, level=-1).reset_index()
merge_temp2 = pd.merge(temp12, temp22, on='hadm_id', how='outer').fillna('missing')
merge_temp2
```

```python
commorbd = commobd.rename(columns={'subject': 'subject_id'})
```

```python
# Merge all patient-hospital level information
patient_hosp_level = pd.merge(pd.merge(commorbd, patientsinfo, on=['subject_id']), procdevnts[['subject_id', 'hadm_id', 'patientweight']].drop_duplicates(['subject_id', 'hadm_id']), on=['subject_id', 'hadm_id'])
```

```python
patient_hosp_level.to_csv("data/patient_hosp_level_info.csv")
```

```python
# merged_lab.csv is the data merged procedure timespan with lab events testing
merged_lab = pd.read_csv("data/merged_lab.csv", index_col=0)
merged_lab.head()
```

```python
# Unstack merged_lab
unstack_lab = merged_lab.set_index(['hadm_id', 'starttime', 'label'])[['maximum', 'minimum']].unstack(level=-1)
unstack_lab.columns = unstack_lab.columns.map('_'.join).str.strip('_')
unstack_lab = unstack_lab.reset_index()
unstack_lab.head()
```

```python
unstack_lab.to_csv("data/unstacked_lab.csv")
```

```python
temp = pd.merge(unstack_lab, patient_hosp_level, on=['hadm_id'])
temp.head()
```

```python
# Dataset without location & location category
dataset = pd.merge(temp, label, on=['subject_id', 'hadm_id']).drop_duplicates('hadm_id')
dataset = dataset.reset_index().drop(columns=['index', 'maximum_Intubated', 'minimum_Intubated'])
dataset
```

```python
dataset.to_csv("data/dataset.csv")
```

```python
# Dataset with all location category
dataset2 = pd.merge(temp, merge_temp, on=['hadm_id'])
dataset2 = pd.merge(dataset2, label, on=['subject_id', 'hadm_id']).drop_duplicates('hadm_id')
dataset2 = dataset2.reset_index().drop(columns=['index', 'maximum_Intubated', 'minimum_Intubated', 'starttime'])
dataset2
```

```python
dataset2.to_csv('data/dataset_with_all_loc.csv')
```

```python
# Dataset with merged location category
dataset3 = pd.merge(temp, merge_temp2, on=['hadm_id'])
dataset3 = pd.merge(dataset3, label, on=['subject_id', 'hadm_id']).drop_duplicates('hadm_id')
dataset3 = dataset3.reset_index().drop(columns=['index', 'maximum_Intubated', 'minimum_Intubated', 'starttime'])
dataset3
```

```python
dataset3.to_csv('data/dataset_with_merged_loc.csv')
```
